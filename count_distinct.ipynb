## COUNT(DISTINCT column) with BlazingSQL [v0.12](https://github.com/BlazingDB/blazingsql/releases/tag/v0.12.0)

```python
from blazingsql import BlazingContext
# connect to BlazingSQL
bc = BlazingContext()
```

#### Download Data
The below cell will check if you have the data for this demo and download it for you if not.

```python
import os
import urllib

# download taxi data
base_url = 'https://raw.githubusercontent.com/BlazingDB/bsql-demos/master/data/'
fn = 'Music.csv'
# check if we already have the file
if not os.path.isfile(f'data/{fn}'):
    # we don't let me know we're downloading it now
    print(f'Downloading {base_url + fn} to data/{fn}')
    # download file
    urllib.request.urlretrieve(base_url + fn, f'data/{fn}')
# we already have data
else:
    # let us know
    print(f'{fn} already downloaded')
    
# identify path to data
cwd = os.getcwd()
data_path = f'{cwd}/data/Music.csv'
```

```python
# create table from CSV
names = ['artist', 'rating', 'year', 'location', 'festival_set']
bc.create_table('music', data_path, header=0, names=names)
```

```python
bc.sql('select COUNT(artist) from music')
```

```python
# New Feature: Ability to do SELECT COUNT (DISTINCT column)
bc.sql('select COUNT(DISTINCT artist) from music')
```

```python
bc.sql('select DISTINCT(artist) from music')
```

```python
bc.sql('select COUNT(DISTINCT rating) from music')
```

```python
bc.sql('select COUNT(DISTINCT location) from music')
```
